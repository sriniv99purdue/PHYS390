```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cmath
%matplotlib inline
```

```python
np.random.seed(100)
mean=0
stddev=1
np_hist = np.random.normal(mean, stddev, size=1000)
```

```python
plt.hist(np_hist,bins=10,edgeColor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('continuous values histogram')
plt.show()
```

```python
random_values = [15, 20, 25, 25, 30, 30, 34, 34, 34, 35, 35, 35, 38, 38, 42, 42, 45,  50]
```

```python
plt.hist(random_values, 20,edgeColor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('discrete values histogram')
plt.show()
```

Specify the number of bins to be used

```python
DataPoints=len(random_values)
```

```python
DataPoints
```

```python
SQDataPoints =int(cmath.sqrt(DataPoints).real)
```

```python
SQDataPoints
```

```python
plt.hist(random_values, bins=SQDataPoints ,edgeColor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('discrete values histogram')
plt.show()
```

```python
plt.hist(random_values, bins=SQDataPoints,log=True, edgeColor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('discrete values histogram')
plt.show()
```
